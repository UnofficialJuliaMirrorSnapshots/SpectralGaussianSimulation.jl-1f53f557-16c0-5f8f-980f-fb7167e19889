# Problem definition

```julia
using GeoStats

problem = SimulationProblem(RegularGrid{Float64}(100,100), :z => Float64, 3)
```

# Solver options

The user can specify the variogram model and the mean of the Gaussian field:

```julia
using SpectralGaussianSimulation

solver = SpecGaussSim(:z => (variogram=GaussianVariogram(range=10.),))

solution = solve(problem, solver)
```

```julia
using Plots

plot(solution, size=(950,200))
```
